##Iris Dataset Classification

```python
## importing pandas
import sklearn
import numpy
import pandas as pd
from sklearn import datasets
```

## Problem Understanding
- choose the correct flowers

## Data Understanding

```python
# dataset in sklearn
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris_data = pd.DataFrame(iris.data,columns=iris.feature_names);
```

```python
iris_data['lable'] = iris.target;
```

```python
iris_data
```

## 3. Data Preprocessing

```python
iris_data.dropna(axis=0,inplace=True)
```

```python
X = iris_data.iloc[:,0:-1]; ##features
Y = iris_data.iloc[:,-1]; ## labels
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=5);
```

```python
X_train
```

```python
Y_train
```

## Modelling

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train,Y_train);
```

```python
model
```

## 5. Evaluation

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
# Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix;
print(confusion_matrix(Y_test,y_predict))
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
print('Accuracy Score',accuracy_score(Y_test, y_predict)*100,'%')
print('Precision Macro Score',precision_score(Y_test, y_predict,average = 'macro')*100,'%')
print('Recall_Score',recall_score(Y_test, y_predict, average = 'macro')*100,'%')
print('F_Score',f1_score(Y_test, y_predict, average = 'macro')*100,'%')
```
